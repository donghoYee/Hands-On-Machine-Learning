```python
import numpy as np

X = 2* np.random.rand(100, 1)
print(X.shape)
```

```python
y = 4 + 3 * X + np.random.randn(100, 1)
```

```python
print(np.random.randn(100, 1). shape)
```

```python
import matplotlib.pyplot as plt
```

```python

plt.scatter(X.reshape(100), y.reshape(100))
plt.show()
```

```python
X_b = np.c_[np.ones((100,1)), X]
print(X_b.shape)
print(X_b[0])
theta_best = np.linalg.inv(X_b.T.dot(X_b)).dot(X_b.T).dot(y)
print(theta_best)
```

```python
plt.scatter(X.reshape(100), y.reshape(100))
print(theta_best.shape)
plt.plot(np.arange(0,2, 0.01), theta_best[0][0] + theta_best[1][0] * np.arange(0,2, 0.01))
plt.show()
```

```python
X_new = np.array([[0],[2]])
X_new_b = np.c_[np.ones((2,1)), X_new]
y_predict = X_new_b.dot(theta_best)
```

```python
print(y_predict)
print(y_predict.shape)
```

```python
plt.plot(X_new, y_predict, "r-")
plt.plot(X, y, "b.")
plt.show()
```

```python
n_epochs = 50
t0, t1 = 5, 50 #파라미터!

def learning_schedule(t):
    return t0 / (t1+t)   #what??

theta = np.random.randn(2, 1)

for epoch in range(n_epochs):
    for i in range(m):
        random_index = np.random.randint(m) #why? -> to pick one in samples(random from 0 to m-1)
        xi = X_b[random_index:random_index+1] #what?
        yi = y[random_index:random_index+1]
        
        gradients = 2 * xi.T.dot(xi.dot(theta) - yi)
        eta = learning_schedule(epoch * m + i)
        theta = theta - eta * gradients
```
