```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import tensorflow as tf
from tensorflow import keras
```

```python
(X_train_full, y_train_full), (X_test, y_test) = keras.datasets.cifar10.load_data()
```

```python
X_train_full = X_train_full / 255.0
X_test = X_test / 255.0
```

```python
from sklearn.model_selection import train_test_split
X_train, X_val, y_train, y_val = train_test_split(X_train_full, y_train_full,test_size=0.2)
```

```python
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
plt.imshow(X_train[3])
```

### 100개의 뉴런을 가진 은닉층 20

```python
input_layer = keras.layers.Input(shape=X_train[0].shape)
x = keras.layers.Flatten()(input_layer)
x = keras.layers.Dense(100, kernel_initializer="he_normal", activation="elu")(x)
x = keras.layers.Dense(100, kernel_initializer="he_normal", activation="elu")(x)
x = keras.layers.Dense(100, kernel_initializer="he_normal", activation="elu")(x)
x = keras.layers.Dense(100, kernel_initializer="he_normal", activation="elu")(x)
x = keras.layers.Dense(100, kernel_initializer="he_normal", activation="elu")(x)
x = keras.layers.Dense(100, kernel_initializer="he_normal", activation="elu")(x)
x = keras.layers.Dense(100, kernel_initializer="he_normal", activation="elu")(x)
x = keras.layers.Dense(100, kernel_initializer="he_normal", activation="elu")(x)
x = keras.layers.Dense(100, kernel_initializer="he_normal", activation="elu")(x)
x = keras.layers.Dense(100, kernel_initializer="he_normal", activation="elu")(x)
x = keras.layers.Dense(100, kernel_initializer="he_normal", activation="elu")(x)
x = keras.layers.Dense(100, kernel_initializer="he_normal", activation="elu")(x)
x = keras.layers.Dense(100, kernel_initializer="he_normal", activation="elu")(x)
x = keras.layers.Dense(100, kernel_initializer="he_normal", activation="elu")(x)
x = keras.layers.Dense(100, kernel_initializer="he_normal", activation="elu")(x)
x = keras.layers.Dense(100, kernel_initializer="he_normal", activation="elu")(x)
x = keras.layers.Dense(100, kernel_initializer="he_normal", activation="elu")(x)
x = keras.layers.Dense(100, kernel_initializer="he_normal", activation="elu")(x)
x = keras.layers.Dense(100, kernel_initializer="he_normal", activation="elu")(x)
x = keras.layers.Dense(100, kernel_initializer="he_normal", activation="elu")(x)
output_layer = keras.layers.Dense(10, activation="softmax")(x)
```

```python
model = keras.Model(inputs=[input_layer], outputs=[output_layer])
```

```python
model.summary()
```

```python
keras.utils.plot_model(model, to_file='model.png', show_shapes=True, show_layer_names=True)
```

```python
checkpoint_cb = keras.callbacks.ModelCheckpoint("model.h5", save_best_only=True)
early_stopping_cb = keras.callbacks.EarlyStopping(patience=5, restore_best_weights=True)
```

```python
optimizer = keras.optimizers.Nadam(learning_rate=0.001, beta_1=0.9, beta_2=0.999)
model.compile(loss="sparse_categorical_crossentropy",
              optimizer=optimizer,
              metrics=["accuracy"])
```

```python
history = model.fit(X_train, y_train, epochs=100,
                    validation_data=(X_val, y_val),
                    callbacks=[checkpoint_cb, early_stopping_cb])
```

```python
pd.DataFrame(history.history).plot(figsize=(8,5))
plt.grid(True)
plt.gca().set_ylim(0,1)
plt.show()
```

### With MCAlphaDropout

```python
class MCAlphaDropout(keras.layers.AlphaDropout):
    def call(self, inputs):
        return self.call(inputs, training=True) ###지금 이거 필요 X (호출할때 다르게 하면 됨)
```

```python
input_layer4 = keras.layers.Input(shape=X_train[0].shape)
x = keras.layers.Flatten()(input_layer4)

x = keras.layers.Dense(100, kernel_initializer="lecun_normal", activation="selu")(x)
x = keras.layers.Dense(100, kernel_initializer="lecun_normal", activation="selu")(x)
x = keras.layers.Dense(100, kernel_initializer="lecun_normal", activation="selu")(x)
x = keras.layers.Dense(100, kernel_initializer="lecun_normal", activation="selu")(x)
x = keras.layers.Dense(100, kernel_initializer="lecun_normal", activation="selu")(x)

x = keras.layers.Dense(100, kernel_initializer="lecun_normal", activation="selu")(x)
x = keras.layers.Dense(100, kernel_initializer="lecun_normal", activation="selu")(x)
x = keras.layers.Dense(100, kernel_initializer="lecun_normal", activation="selu")(x)
x = keras.layers.Dense(100, kernel_initializer="lecun_normal", activation="selu")(x)
x = keras.layers.Dense(100, kernel_initializer="lecun_normal", activation="selu")(x)

x = keras.layers.Dense(100, kernel_initializer="lecun_normal", activation="selu")(x)
x = keras.layers.Dense(100, kernel_initializer="lecun_normal", activation="selu")(x)
x = keras.layers.Dense(100, kernel_initializer="lecun_normal", activation="selu")(x)
x = keras.layers.Dense(100, kernel_initializer="lecun_normal", activation="selu")(x)
x = keras.layers.Dense(100, kernel_initializer="lecun_normal", activation="selu")(x)

x = keras.layers.Dense(100, kernel_initializer="lecun_normal", activation="selu")(x)
x = keras.layers.Dense(100, kernel_initializer="lecun_normal", activation="selu")(x)
x = keras.layers.Dense(100, kernel_initializer="lecun_normal", activation="selu")(x)
x = keras.layers.Dense(100, kernel_initializer="lecun_normal", activation="selu")(x)
x = keras.layers.AlphaDropout(rate=0.3)(x,training=True)
x = keras.layers.Dense(100, kernel_initializer="lecun_normal", activation="selu")(x)

x = keras.layers.Dense(100, kernel_initializer="lecun_normal", activation="selu")(x)
x = keras.layers.AlphaDropout(rate=0.3)(x,training=True)
x = keras.layers.Dense(100, kernel_initializer="lecun_normal", activation="selu")(x)
x = keras.layers.Dense(100, kernel_initializer="lecun_normal", activation="selu")(x)
x = keras.layers.AlphaDropout(rate=0.3)(x,training=True)
x = keras.layers.Dense(100, kernel_initializer="lecun_normal", activation="selu")(x)
x = keras.layers.Dense(100, kernel_initializer="lecun_normal", activation="selu")(x)

output_layer4 = keras.layers.Dense(10, activation="softmax")(x)
```

```python
model4 = keras.Model(inputs=[input_layer4], outputs=[output_layer4])
keras.utils.plot_model(model4, to_file='model4.png', show_shapes=True, show_layer_names=True)
```

```python
model4.compile(loss="sparse_categorical_crossentropy",
              optimizer=optimizer,
              metrics=["accuracy"])
checkpoint_cb = keras.callbacks.ModelCheckpoint("model4.h5", save_best_only=True)
history4 = model4.fit(X_train, y_train, epochs=100,
                    validation_data=(X_val, y_val),
                    callbacks=[checkpoint_cb, early_stopping_cb])
```

```python
pd.DataFrame(history4.history).plot(figsize=(8,5))
plt.grid(True)
plt.gca().set_ylim(0,1)
plt.show()
```

## With 1CycleScheduling

```python
K = keras.backend
def one_cycle_scheduler_fn(epoch): #n_epoch=100 가정
    max_lr = 0.01
    start_lr = 0.001
    end_lr = 0.0001
    if(epoch < 30):
        return start_lr + epoch * (max_lr - start_lr) / 30 
    if(epoch < 60):
        return max_lr - (epoch-30) * (max_lr - start_lr) / 30 
    return start_lr - (epoch-60) * (start_lr - end_lr) / 30 
```

```python
epoches = np.arange(100)
plt.plot(epoches, [one_cycle_scheduler_fn(i) for i in epoches])
```

```python
lr_scheduler = keras.callbacks.LearningRateScheduler(one_cycle_scheduler_fn)
```

```python
model5 = keras.Model(inputs=[input_layer4], outputs=[output_layer4])
keras.utils.plot_model(model5, to_file='model5.png', show_shapes=True, show_layer_names=True)
```

```python
model5.compile(loss="sparse_categorical_crossentropy",
              optimizer=optimizer,
              metrics=["accuracy"])
checkpoint_cb = keras.callbacks.ModelCheckpoint("model5.h5", save_best_only=True)
history5 = model5.fit(X_train, y_train, epochs=100,
                    validation_data=(X_val, y_val),
                    callbacks=[lr_scheduler])
```

```python
pd.DataFrame(history5.history).plot(figsize=(8,5))
plt.gca().set_ylim(0,1)
plt.show()
```
