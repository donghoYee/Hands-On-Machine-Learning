# Train With MNIST

```python
import tensorflow as tf
from tensorflow import keras 
import numpy as np
```

```python
from keras.datasets.mnist import load_data
(X_train_full, y_train_full), (X_test, y_test)= load_data()

X_train_full = X_train_full.reshape(-1, 28*28)
X_test = X_test.reshape(-1,28*28)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_val, y_train, y_val = train_test_split(X_train_full, y_train_full)
```

```python
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
X_train = X_train.astype("float")/255
X_test = X_test.astype("float")/255
X_val = X_val.astype("float")/255
```

```python
X_train[0]
```

```python
# use Randomized_search cv!
```

```python
def build_model(n_hidden=1, n_neurons=300, lr = 0.001, input_shape=[784]):
    model = keras.models.Sequential()
    model.add(keras.layers.InputLayer(input_shape=input_shape))
    
    for layer in range(n_hidden):
        model.add(keras.layers.Dense(n_neurons, activation="relu"))
    
    model.add(keras.layers.Dense(10, activation="softmax"))
    optimizer = keras.optimizers.Adam(learning_rate=lr)
    model.compile(loss="sparse_categorical_crossentropy", optimizer=optimizer, metrics=["accuracy"]) #for y with numbers(not one-hot-vectors)
    
    return model
  
```

```python
keras_clf = keras.wrappers.scikit_learn.KerasClassifier(build_model)
```

```python
early_stopping_cb = keras.callbacks.EarlyStopping(patience=10)
checkpoint_cb = keras.callbacks.ModelCheckpoint("mnist_model.h5", save_best_only=True)
```

```python
from sklearn.model_selection import RandomizedSearchCV

params = {
    "n_hidden" : [1,2,3,4],
    "n_neurons": [300,400,500],
    "lr": np.linspace(0.001, 0.1, 10)
}

rnd_search_cv = RandomizedSearchCV(keras_clf, params, n_iter=10, cv=3) # make models accordingly
```

```python
rnd_search_cv.fit(X_train, y_train, validation_data=(X_val, y_val), epochs=100, callbacks=[ checkpoint_cb])
```

```python
best_model = rnd_search_cv.best_estimator_
print(rnd_search_cv.best_score_)
```

```python
y_pred = best_model.predict(X_test)
from sklearn.metrics import accuracy_score
print("accuracy: ", accuracy_score(y_test, y_pred))
```
