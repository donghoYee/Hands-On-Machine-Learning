```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import tensorflow as tf
from tensorflow import keras
```

```python
# make a function that accepts paramitors about network and compile that network and returns it!
def build_model(n_hidden=1, n_neurons=30, learning_rate=3e-3, input_shape=[8]):
    model = keras.models.Sequential() # make a wrapper!
    model.add(keras.layers.InputLayer(input_shape=input_shape))
    
    for layer in range(n_hidden):
        model.add(keras.layers.Dense(n_neurons, activation="relu"))
    
    model.add(keras.layers.Dense(1)) # no activation func and just outputs one data for regression!
    
    optimizer = keras.optimizers.SGD(learning_rate=learning_rate)
    model.compile(loss="mse", optimizer=optimizer)
    return model
```

```python
#we can use keras.wrappers.scikit_learn.KerasRegressor to make this function into a scikit-learn model!
keras_reg = keras.wrappers.scikit_learn.KerasRegressor(build_model)
```

```python
from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

housing = fetch_california_housing()
X_train_full, X_test, y_train_full, y_test = train_test_split(housing.data, housing.target)
X_train, X_valid, y_train, y_valid = train_test_split(X_train_full, y_train_full)

scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
X_valid = scaler.transform(X_valid)
```

```python
keras_reg.fit(X_train, y_train, epochs=100, validation_data=(X_valid, y_valid), callbacks=[keras.callbacks.EarlyStopping(patience=10)])
```

```python
mse_test = keras_reg.score(X_test, y_test) #we can use this as if it were a scikit-learn regressor!
```

```python
X_new = X_test[:3]
y_pred = keras_reg.predict(X_new)
print(y_pred)
print(y_test[:3])
```

## Use RandomizedSearchCV to determine hyperparamitors for keras models!

```python
from scipy.stats import reciprocal
from sklearn.model_selection import RandomizedSearchCV

param_distribs = {
    "n_hidden": [0,1,2,3], # we only use one model so param's names can be like this!
    "n_neurons": np.arange(1, 100),
    "learning_rate": reciprocal(3e-4, 3e-2),
}

rnd_search_cv = RandomizedSearchCV(keras_reg, param_distribs, n_iter=10, cv=3)
rnd_search_cv.fit(X_train, y_train, epochs=100, validation_data=(X_valid, y_valid),
                  callbacks=[keras.callbacks.EarlyStopping(patience=10)])
    
```

```python
rnd_search_cv.best_params_
```

```python
rnd_search_cv.best_score_
```

```python
best_model = rnd_search_cv.best_estimator_
```

```python
import pickle
import joblib
joblib.dump(best_model, "RandomizedSearchCV_best_model.pkl")
```
